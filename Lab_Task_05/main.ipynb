```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image = cv2.imread('F:\Assigmnt\AI LAB\Lab_Task_05\pic.jpg')


image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
print("Shape of the image:", image.shape) 
print("Size of the image:", image.size)
print("Type of the image:", image.dtype)
```

```python

cropped_image = image[50:250, 100:300]


cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)
plt.imshow(cropped_image_rgb)
plt.axis('off')
plt.show()
```

```python
# Draw a red rectangle
cv2.rectangle(image, (50, 50), (200, 200), (0, 0, 255), 3)

# Draw a blue circle
cv2.circle(image, (400, 400), 50, (255, 0, 0), -1)

# Show the result
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply simple thresholding
_, threshold_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Display thresholded image
plt.imshow(threshold_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)

# Display edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

# Display blurred image
blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)
plt.imshow(blurred_image_rgb)
plt.axis('off')
plt.show()
```

```python
# Find contours in the image
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)

# Display the image with contours
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Load the pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Draw rectangles around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Display the result
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Release the webcam
# Open webcam (0 is the default camera)
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert the frame to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Display the resulting frame
    plt.imshow(frame_rgb)
    plt.axis('off')
    plt.show()

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam
cap.release()
cv2.destroyAllWindows()
```
